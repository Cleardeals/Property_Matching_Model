```python
import pandas as pd
import numpy as np
import re
import json
import time
import os
from geopy.geocoders import Nominatim
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np
import re
import json
import time
import os
from geopy.geocoders import Nominatim
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np
import re
import json
import time
import os
from geopy.geocoders import Nominatim
from sklearn.cluster import KMeans

def get_or_create_geo_mapping(df):
    """
    Loads geocoding from a file or creates it by calling an API.
    """
    json_path = 'geo_mapping.json'
    if os.path.exists(json_path):
        print(f"‚úÖ Found existing geocoding file at '{json_path}'. Loading...")
        with open(json_path, 'r') as f:
            return json.load(f)

    print(f"‚ö†Ô∏è No geocoding file found. Creating a new one using geopy...")
    print("This will take several minutes as it respects API rate limits.")

    geolocator = Nominatim(user_agent="property_matcher_app_v6")
    unique_locations = df['Location'].dropna().unique()
    geo_mapping = {}

    city_prefix_map = {
        'A': 'Ahmedabad', 'P': 'Pune', 'G': 'Gandhinagar',
        'S': 'Surat', 'V': 'Vadodara', 'B': 'Vadodara'
    }

    for location_str in unique_locations:
        if not isinstance(location_str, str) or '-' not in location_str:
            continue

        parts = location_str.split('-', 1)
        prefix = parts[0].strip()
        area = parts[1].strip()
        city = city_prefix_map.get(prefix, 'Ahmedabad')

        query = f"{area}, {city}, India"
        try:
            location_data = geolocator.geocode(query, timeout=10)
            if location_data:
                geo_mapping[location_str] = (location_data.latitude, location_data.longitude)
                print(f"‚úÖ Found: {query} -> ({location_data.latitude:.4f}, {location_data.longitude:.4f})")
            else:
                print(f"‚ö†Ô∏è Not Found: {query}")
            time.sleep(1)
        except Exception as e:
            print(f"‚ùå Error for '{query}': {e}")

    with open(json_path, 'w') as f:
        json.dump(geo_mapping, f, indent=4)
    print(f"‚úÖ Geocoding complete. Saved to '{json_path}'.")
    return geo_mapping

def run_data_pipeline():
    """
    Loads, cleans, and preprocesses the property data, including location clustering
    and calculating a three-state status (Active, Expired, Sold).
    """
    try:
        property_df = pd.read_csv('PropertyData.csv', low_memory=False)
        print("‚úÖ Successfully loaded PropertyData.csv")
    except FileNotFoundError:
        print("‚ùå Error: 'PropertyData.csv' not found.")
        return None, None, None, None

    df = property_df.copy()
    # Ensure necessary columns are included
    features_to_use = [
        'BHK', 'Property-Price', 'City1', 'Property-On-Floor', 'Property-Facing',
        'Age-Of-Property', 'Super-Built-up-Construction-Area', 'Carpet-Construction-Area',
        'Bathroom', 'Furniture-Details', 'Property-Status', 'Current-Status', 'Location', 'Parking-Details',
        'No-Of-Lift-Per-Block', 'Service_Expiry_Date', 'Tag'
    ]
    df = df[features_to_use]
    print(f"‚úÖ Selected {len(features_to_use)} features for modeling.")

    # --- UPDATED: Three-State Status Calculation ---
    print("\nüóìÔ∏è Calculating property status (Active/Expired/Sold)...")
    
    # Define what constitutes a 'Sold' property
    sold_statuses = ['Sold-CD', 'Sold-Others', 'Rented-CD']
    
    # Define the reference date for checking expiry
    current_date = pd.to_datetime('2025-07-31')
    
    # Convert Service_Expiry_Date to datetime objects
    df['Service_Expiry_Date'] = pd.to_datetime(df['Service_Expiry_Date'], errors='coerce', dayfirst=True)
    
    # Create the new reliable status column with three states using np.select
    conditions = [
        df['Property-Status'].isin(sold_statuses),
        df['Service_Expiry_Date'] < current_date
    ]
    choices = ['Sold', 'Expired']
    df['Calculated_Status'] = np.select(conditions, choices, default='Active')
    
    print(f"‚úÖ Status calculation complete. Status distribution:")
    print(df['Calculated_Status'].value_counts())

    geo_mapping = get_or_create_geo_mapping(df)
    median_lat = np.median([val[0] for val in geo_mapping.values() if val and val[0] is not None])
    median_lon = np.median([val[1] for val in geo_mapping.values() if val and val[1] is not None])

    df[['Latitude', 'Longitude']] = df['Location'].apply(
        lambda x: pd.Series(geo_mapping.get(str(x), (median_lat, median_lon)))
    )

    print("\nüìç Creating geographical clusters using K-Means...")
    coords = df[['Latitude', 'Longitude']]
    kmeans = KMeans(n_clusters=30, random_state=42, n_init='auto')
    df['Location_Cluster'] = kmeans.fit_predict(coords)
    print(f"‚úÖ Created {df['Location_Cluster'].nunique()} unique location clusters.")

    print("\n‚öôÔ∏è Starting data cleaning process...")
    def clean_price(price):
        if not isinstance(price, str): return np.nan
        price_str = price.lower()
        try:
            numbers = re.findall(r'[\d\.]+', price_str)
            if not numbers: return np.nan
            value = float(numbers[0])
            if 'cr' in price_str: return value * 100
            return value
        except (ValueError, IndexError): return np.nan

    def clean_area(area):
        if not isinstance(area, str): return np.nan
        area_str = area.lower()
        try:
            numbers = re.findall(r'[\d\.]+', area_str)
            if not numbers: return np.nan
            value = float(numbers[0])
            if 'yard' in area_str: return value * 9
            return value
        except (ValueError, IndexError): return np.nan

    def clean_floor(floor):
        if not isinstance(floor, str): return np.nan
        floor_str = floor.lower().replace('g', '0')
        try:
            numbers = re.findall(r'\d+', floor_str)
            if numbers: return int(numbers[0])
            return np.nan
        except (ValueError, IndexError): return np.nan

    def clean_age(age):
        if not isinstance(age, str): return np.nan
        age_str = age.lower()
        if 'new' in age_str or 'under' in age_str: return 0
        try:
            numbers = [int(s) for s in re.findall(r'\d+', age_str)]
            if numbers: return sum(numbers) / len(numbers)
            return np.nan
        except (ValueError, IndexError): return np.nan

    df.replace('-', np.nan, inplace=True)
    df['Property-Price'] = df['Property-Price'].apply(clean_price)
    df['Super-Built-up-Construction-Area'] = df['Super-Built-up-Construction-Area'].apply(clean_area)
    df['Carpet-Construction-Area'] = df['Carpet-Construction-Area'].apply(clean_area)
    df['Property-On-Floor'] = df['Property-On-Floor'].apply(clean_floor)
    df['Age-Of-Property'] = df['Age-Of-Property'].apply(clean_age)
    df['BHK'] = pd.to_numeric(df['BHK'].astype(str).str.extract(r'(\d+)', expand=False), errors='coerce')
    df['Bathroom'] = pd.to_numeric(df['Bathroom'], errors='coerce')
    df['No-Of-Lift-Per-Block'] = pd.to_numeric(df['No-Of-Lift-Per-Block'], errors='coerce')
    df.loc[df['Bathroom'] > 20, 'Bathroom'] = np.nan

    numerical_cols = ['Property-Price', 'Super-Built-up-Construction-Area', 'Carpet-Construction-Area',
                      'Property-On-Floor', 'Age-Of-Property', 'BHK', 'Bathroom', 'No-Of-Lift-Per-Block']
    for col in numerical_cols:
        df[col].fillna(df[col].median(), inplace=True)

    categorical_cols = ['City1', 'Property-Facing', 'Furniture-Details', 'Property-Status',
                        'Current-Status', 'Parking-Details', 'Location_Cluster', 'Calculated_Status']
    for col in categorical_cols:
        df[col] = df[col].astype(str)
        df[col].fillna(df[col].mode()[0], inplace=True)

    print("‚úÖ Data cleaning and imputation complete.")
    
    # Create Preprocessor for use in later steps
    numerical_features = [col for col in numerical_cols]
    # We exclude our new Calculated_Status from the training features
    categorical_features = [col for col in categorical_cols if col != 'Calculated_Status']
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', MinMaxScaler(), numerical_features),
            ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
        ],
        remainder='drop'
    )
    
    return df, numerical_cols, categorical_cols, preprocessor

    
# --- Block 2: Exploratory Data Analysis (EDA) ---
def perform_eda(df_cleaned):
    print("\n\n--- üî¨ Starting Exploratory Data Analysis ---")
    print("\n--- üìä Descriptive Statistics ---")
    print(df_cleaned.describe())
    sns.set_style("whitegrid")
    print("\n--- üìà Visualizing Numerical Distributions ---")
    df_cleaned.hist(bins=30, figsize=(20, 15), layout=(4, 3))
    plt.suptitle('Distribution of Numerical Features', size=20, y=1.02)
    plt.tight_layout()
    plt.show()

# Block 3- Training the model
def train_autoencoder(df_cleaned, preprocessor):
    """
    Builds and trains the tuned autoencoder model using the provided preprocessor.
    """
    print("\n\n--- ü§ñ Preparing data for the model ---")
    
    # Use the preprocessor that was created in the data pipeline
    X_processed = preprocessor.fit_transform(df_cleaned)
    print(f"‚úÖ Preprocessing complete. Shape of model input data: {X_processed.shape}")

    input_dim = X_processed.shape[1]
    embedding_dim = 64

    # --- Model Architecture ---
    input_layer = Input(shape=(input_dim,))
    encoder = Dense(256)(input_layer)
    encoder = LeakyReLU()(encoder)
    encoder = Dense(128)(encoder)
    encoder = LeakyReLU()(encoder)
    
    bottleneck = Dense(embedding_dim)(encoder)
    bottleneck = LeakyReLU()(bottleneck)

    decoder = Dense(128)(bottleneck)
    decoder = LeakyReLU()(decoder)
    decoder = Dense(256)(decoder)
    decoder = LeakyReLU()(decoder)
    output_layer = Dense(input_dim, activation='sigmoid')(decoder)

    autoencoder = Model(inputs=input_layer, outputs=output_layer)
    encoder_model = Model(inputs=input_layer, outputs=bottleneck)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')
    print("\n‚úÖ Tuned autoencoder model built.")
    autoencoder.summary()

    print("\n--- üèãÔ∏è‚Äç‚ôÇÔ∏è Training the Autoencoder ---")
    autoencoder.fit(
        X_processed, X_processed,
        epochs=100,
        batch_size=32,
        shuffle=True,
        validation_split=0.1,
        verbose=1
    )
    print("‚úÖ Model training complete.")
    return encoder_model, autoencoder, X_processed
    
# BLOCK - 4 
def find_similar_properties(encoder_model, autoencoder, X_processed, df_cleaned):
    """
    Uses the trained encoder to find and display similar properties,
    and now returns the embeddings for further analysis.
    """
    print("\n\n--- üéØ Finding Similar Properties ---")
    
    loss = autoencoder.evaluate(X_processed, X_processed, verbose=0)
    print(f"\n--- üìà Model Evaluation ---")
    print(f"Final Mean Squared Error (Reconstruction Loss): {loss:.6f}")
    
    all_embeddings = encoder_model.predict(X_processed)

    active_mask = df_cleaned['Calculated_Status'] == 'Active'
    query_mask = df_cleaned['Calculated_Status'] == 'Expired'

    active_indices = df_cleaned[active_mask].index
    query_indices = df_cleaned[query_mask].index
    
    active_embeddings = all_embeddings[active_indices]

    print(f"\nFound {len(active_indices)} active properties to match against.")
    print(f"Found {len(query_indices)} expired properties to find matches for.")

    if len(active_indices) > 0 and len(query_indices) > 0:
        K = 10 
        knn = NearestNeighbors(n_neighbors=K, metric='cosine')
        knn.fit(active_embeddings)

        query_original_index = np.random.choice(query_indices)
        query_embedding = all_embeddings[query_original_index].reshape(1, -1)

        print("\n--- Example Match ---")
        print(f"Finding similar properties for EXPIRED property at index: {query_original_index}")
        print("Query Property Details:")
        query_property = df_cleaned.iloc[query_original_index]
        print(query_property)
        print("-" * 20)

        distances, indices = knn.kneighbors(query_embedding)

        print(f"Top {K} initial matches found. Filtering for price relevance...")
        
        query_price = query_property['Property-Price']
        price_tolerance = 0.50
        lower_bound = query_price * (1 - price_tolerance)
        upper_bound = query_price * (1 + price_tolerance)
        
        filtered_matches = []
        for i, index in enumerate(indices[0]):
            original_df_index = active_indices[index]
            match_price = df_cleaned.iloc[original_df_index]['Property-Price']
            
            if lower_bound <= match_price <= upper_bound:
                similarity_score = 1 - distances[0][i]
                filtered_matches.append((original_df_index, similarity_score))

        print(f"\n‚úÖ Found {len(filtered_matches)} matches within a {price_tolerance*100}% price range.")
        print(f"--- Top {min(5, len(filtered_matches))} Financially Relevant ACTIVE Properties ---")
        
        if not filtered_matches:
            print("No active properties found within the price range.")
        else:
            for i, (original_df_index, similarity_score) in enumerate(filtered_matches[:5]):
                print(f"\nRank {i+1}: Property at index {original_df_index} (Similarity: {similarity_score:.4f})")
                print(df_cleaned.iloc[original_df_index])
    else:
        print("\nCould not perform matching: no active or expired properties found to process.")
        
    # --- ADD THIS LINE ---
    return all_embeddings
```

```python
import pandas as pd
import numpy as np
import json
from sklearn.neighbors import NearestNeighbors

def create_all_matches_robust(encoder_model, X_processed, df_cleaned):
    """
    Finds and saves the best active matches for every single expired property,
    using the 'Tag' column as the primary key with robust index mapping.
    """
    print("üöÄ Starting the robust bulk matching process for all expired properties...")
    
    # Create a clean copy of the dataframe with reset index to ensure alignment
    df_work = df_cleaned.reset_index(drop=True).copy()
    
    # Verify that X_processed and df_work have the same number of rows
    if len(X_processed) != len(df_work):
        print(f"‚ùå ERROR: Mismatch between processed data ({len(X_processed)}) and dataframe ({len(df_work)}) rows!")
        return
    
    # Generate embeddings
    print("üîÑ Generating embeddings...")
    all_embeddings = encoder_model.predict(X_processed)
    
    # Create a mapping from array position to Tag for easy lookup
    position_to_tag = df_work['Tag'].tolist()
    tag_to_position = {tag: pos for pos, tag in enumerate(position_to_tag)}
    
    # Create separate lists for active and expired properties with their positions
    active_data = []
    expired_data = []
    
    for pos, row in df_work.iterrows():
        property_data = {
            'position': pos,
            'tag': row['Tag'],
            'embedding': all_embeddings[pos],
            'price': row['Property-Price'],
            'status': row['Calculated_Status']
        }
        
        if row['Calculated_Status'] == 'Active':
            active_data.append(property_data)
        elif row['Calculated_Status'] == 'Expired':
            expired_data.append(property_data)
    
    print(f"‚úÖ Found {len(active_data)} active properties and {len(expired_data)} expired properties")
    
    if len(active_data) == 0 or len(expired_data) == 0:
        print("‚ùå Could not create matches. No active or expired properties found.")
        return
    
    # Prepare embeddings for KNN
    active_embeddings = np.array([prop['embedding'] for prop in active_data])
    active_tags = [prop['tag'] for prop in active_data]
    active_prices = [prop['price'] for prop in active_data]
    
    # Train KNN model
    print("üîÑ Training KNN model...")
    knn = NearestNeighbors(n_neighbors=min(10, len(active_data)), metric='cosine')
    knn.fit(active_embeddings)
    
    # Process each expired property
    expired_to_active_matches = {}
    price_tolerance = 0.5
    
    print(f"üîÑ Processing {len(expired_data)} expired properties...")
    
    for i, expired_prop in enumerate(expired_data):
        if i % 100 == 0:  # Progress indicator
            print(f"  Processed {i}/{len(expired_data)} properties...")
        
        query_embedding = expired_prop['embedding'].reshape(1, -1)
        query_tag = expired_prop['tag']
        query_price = expired_prop['price']
        
        # Find nearest neighbors
        distances, indices = knn.kneighbors(query_embedding)
        
        # Calculate price bounds
        lower_bound = query_price * (1 - price_tolerance)
        upper_bound = query_price * (1 + price_tolerance)
        
        # Filter matches by price and collect tags
        final_match_tags = []
        match_details = []
        
        for j, neighbor_idx in enumerate(indices[0]):
            active_tag = active_tags[neighbor_idx]
            active_price = active_prices[neighbor_idx]
            similarity_score = 1 - distances[0][j]
            
            if lower_bound <= active_price <= upper_bound:
                final_match_tags.append(active_tag)
                match_details.append({
                    'tag': active_tag,
                    'similarity': float(similarity_score),
                    'price': active_price
                })
            
            if len(final_match_tags) >= 5:
                break
        
        # Store matches using Tag as key
        expired_to_active_matches[query_tag] = {
            'matched_tags': final_match_tags,
            'details': match_details,
            'query_price': query_price
        }
    
    print(f"‚úÖ Matching complete! Processed {len(expired_data)} expired properties")
    
    # Save the matches to files
    # Simple version (just tags)
    simple_matches = {k: v['matched_tags'] for k, v in expired_to_active_matches.items()}
    with open('matches.json', 'w') as f:
        json.dump(simple_matches, f, indent=4)
    
    # Detailed version (with similarity scores and prices)
    with open('matches_detailed.json', 'w') as f:
        json.dump(expired_to_active_matches, f, indent=4)
    
    print("üìÅ Results saved to:")
    print("  - 'matches.json' (simple tag-to-tags mapping)")
    print("  - 'matches_detailed.json' (includes similarity scores and prices)")
    
    # Validation check
    print("\nüîç Validation Results:")
    total_matches = sum(len(matches['matched_tags']) for matches in expired_to_active_matches.values())
    properties_with_matches = sum(1 for matches in expired_to_active_matches.values() if matches['matched_tags'])
    
    print(f"  - Properties with at least one match: {properties_with_matches}/{len(expired_data)}")
    print(f"  - Total matches created: {total_matches}")
    print(f"  - Average matches per property: {total_matches/len(expired_data):.2f}")
    
    return expired_to_active_matches


def validate_matches(df_cleaned, matches_file='matches.json'):
    """
    Validates that the matches are correctly mapped using Tags
    """
    print("üîç Validating matches against original data...")
    
    # Load matches
    with open(matches_file, 'r') as f:
        matches = json.load(f)
    
    # Create tag to data mapping for quick lookup
    tag_to_data = df_cleaned.set_index('Tag').to_dict('index')
    
    validation_results = []
    
    for expired_tag, matched_tags in list(matches.items())[:5]:  # Check first 5 for demo
        if expired_tag not in tag_to_data:
            print(f"‚ö†Ô∏è  Expired tag '{expired_tag}' not found in original data")
            continue
            
        expired_data = tag_to_data[expired_tag]
        print(f"\n--- Validating matches for Expired Property: {expired_tag} ---")
        print(f"Expired Property - Price: {expired_data['Property-Price']}, BHK: {expired_data['BHK']}")
        
        valid_matches = 0
        for match_tag in matched_tags:
            if match_tag not in tag_to_data:
                print(f"‚ö†Ô∏è  Match tag '{match_tag}' not found in original data")
                continue
                
            match_data = tag_to_data[match_tag]
            if match_data['Calculated_Status'] != 'Active':
                print(f"‚ö†Ô∏è  Match {match_tag} is not Active (Status: {match_data['Calculated_Status']})")
                continue
                
            price_diff = abs(expired_data['Property-Price'] - match_data['Property-Price']) / expired_data['Property-Price'] * 100
            print(f"  ‚úÖ Match: {match_tag} - Price: {match_data['Property-Price']} (Diff: {price_diff:.1f}%), BHK: {match_data['BHK']}")
            valid_matches += 1
        
        validation_results.append({
            'expired_tag': expired_tag,
            'total_matches': len(matched_tags),
            'valid_matches': valid_matches
        })
    
    print(f"\nüìä Validation Summary (sample of {len(validation_results)} properties):")
    for result in validation_results:
        print(f"  {result['expired_tag']}: {result['valid_matches']}/{result['total_matches']} valid matches")


def get_property_details_by_tag(df_cleaned, tag):
    """
    Helper function to get property details by Tag
    """
    property_data = df_cleaned[df_cleaned['Tag'] == tag]
    if property_data.empty:
        print(f"‚ùå No property found with Tag: {tag}")
        return None
    
    return property_data.iloc[0]


# Usage example:
"""
# After training your model, use this robust matching function:
matches_dict = create_all_matches_robust(encoder_model, X_processed, df_cleaned)

# Validate the results:
validate_matches(df_cleaned, 'matches.json')

# Look up a specific property:
property_details = get_property_details_by_tag(df_cleaned, 'some_tag_here')
"""
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_matches(json_path='matches.json'):
    """
    Analyzes the matching results from the generated JSON file.
    """
    try:
        with open(json_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{json_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{json_path}' not found. Please run the matching script first.")
        return

    if not matches_data:
        print("‚ö†Ô∏è The matches file is empty. No analysis can be performed.")
        return

    # --- 1. Calculate the number of matches for each expired property ---
    match_counts = {expired_tag: len(active_tags) for expired_tag, active_tags in matches_data.items()}
    
    # Convert to a pandas Series for easy analysis
    match_counts_series = pd.Series(match_counts)

    # --- 2. Calculate and print the average number of matches ---
    average_matches = match_counts_series.mean()
    print("\n--- üìä Match Analysis ---")
    print(f"Average number of active matches per expired property: {average_matches:.2f}")

    # --- 3. Analyze the distribution of match counts ---
    match_distribution = match_counts_series.value_counts().sort_index()
    print("\n--- üìà Distribution of Match Counts ---")
    print("This shows how many expired properties have 0 matches, 1 match, 2 matches, etc.")
    print(match_distribution)

    # --- 4. Visualize the distribution ---
    plt.figure(figsize=(10, 6))
    sns.barplot(x=match_distribution.index, y=match_distribution.values, palette='mako')
    plt.title('Distribution of the Number of Matches per Expired Property')
    plt.xlabel('Number of Active Matches Found')
    plt.ylabel('Number of Expired Properties')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# --- How to run this in your notebook ---
# analyze_matches()
```

```python
analyze_matches()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_zero_match_outliers(df_cleaned, matches_path='matches.json'):
    """
    Analyzes the distribution of all key numerical columns for expired properties with zero matches.
    Accepts the cleaned DataFrame directly.
    """
    try:
        with open(matches_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{matches_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{matches_path}' not found. Please run the matching script first.")
        return

    # --- 1. Identify properties with zero matches ---
    zero_match_tags = [tag for tag, matches in matches_data.items() if len(matches) == 0]
    print(f"\nFound {len(zero_match_tags)} expired properties with 0 matches.")

    # --- 2. Filter the main dataframe to get these properties ---
    if 'Tag' not in df_cleaned.columns:
        print("‚ùå Error: 'Tag' column not found in the provided DataFrame.")
        return
        
    zero_match_df = df_cleaned[df_cleaned['Tag'].isin(zero_match_tags)]

    if zero_match_df.empty:
        print("‚ö†Ô∏è No properties found for the identified tags.")
        return

    # --- 3. Analyze and visualize the distribution for each numerical column ---
    numerical_cols_to_analyze = [
        'Property-Price', 'Super-Built-up-Construction-Area', 
        'Carpet-Construction-Area', 'Age-Of-Property', 'BHK', 'Bathroom'
    ]
    
    print("\n--- üìä Distribution Analysis for Properties with 0 Matches ---")

    for col in numerical_cols_to_analyze:
        if col not in zero_match_df.columns:
            print(f"‚ö†Ô∏è Skipping '{col}' as it's not in the DataFrame.")
            continue
            
        plt.figure(figsize=(12, 6))
        sns.histplot(zero_match_df[col], bins=20, kde=True)
        plt.title(f'Distribution of "{col}" for Zero-Match Properties')
        plt.xlabel(col)
        plt.ylabel('Number of Properties')
        
        mean_val = zero_match_df[col].mean()
        plt.axvline(mean_val, color='red', linestyle='--', label=f"Mean: {mean_val:.2f}")
        plt.legend()
        plt.grid(True)
        plt.show()
        
        print(f"\nDescriptive Statistics for '{col}':")
        print(zero_match_df[col].describe())
        print("-" * 40)

# --- How to run this in your notebook ---
# analyze_zero_match_outliers(df_cleaned)
```

```python
analyze_zero_match_outliers(df_cleaned)
```

```python
import json
import pandas as pd
import numpy as np

def analyze_outlier_impact(df_cleaned, matches_path='matches.json'):
    """
    Compares zero-match properties to normal properties to identify outliers.
    """
    try:
        with open(matches_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{matches_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{matches_path}' not found. Please run the matching script first.")
        return

    # --- 1. Separate the two groups of properties ---
    zero_match_tags = [tag for tag, matches in matches_data.items() if len(matches) == 0]
    
    zero_match_df = df_cleaned[df_cleaned['Tag'].isin(zero_match_tags)]
    normal_df = df_cleaned[~df_cleaned['Tag'].isin(zero_match_tags)]
    
    print(f"Analyzing {len(zero_match_df)} zero-match properties against {len(normal_df)} normal properties.")

    # --- 2. Calculate outlier bounds from the 'normal' properties ---
    numerical_cols = [
        'Property-Price', 'Super-Built-up-Construction-Area', 
        'Carpet-Construction-Area', 'Age-Of-Property', 'BHK', 'Bathroom'
    ]
    
    outlier_bounds = {}
    for col in numerical_cols:
        Q1 = normal_df[col].quantile(0.25)
        Q3 = normal_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_bounds[col] = (lower_bound, upper_bound)

    # --- 3. Identify which of the zero-match properties are outliers ---
    outlier_properties = []
    outlier_feature_counts = {col: 0 for col in numerical_cols}

    for index, row in zero_match_df.iterrows():
        is_an_outlier = False
        for col in numerical_cols:
            lower, upper = outlier_bounds[col]
            if not (lower <= row[col] <= upper):
                is_an_outlier = True
                outlier_feature_counts[col] += 1
        
        if is_an_outlier:
            outlier_properties.append(row['Tag'])
            
    # --- 4. Report the findings ---
    num_outliers = len(outlier_properties)
    total_zero_match = len(zero_match_df)
    percentage = (num_outliers / total_zero_match) * 100 if total_zero_match > 0 else 0

    print("\n--- üìä Outlier Analysis Results ---")
    print(f"Out of {total_zero_match} properties with zero matches, {num_outliers} ({percentage:.2f}%) were identified as outliers in at least one numerical feature.")
    
    print("\n--- üî• Most Frequent Outlier Features ---")
    print("This shows which features were most often the reason a property was an outlier:")
    
    # Sort and display the feature counts
    sorted_outlier_features = sorted(outlier_feature_counts.items(), key=lambda item: item[1], reverse=True)
    for feature, count in sorted_outlier_features:
        if count > 0:
            print(f"- {feature}: {count} properties")

# --- How to run this in your notebook ---
# analyze_outlier_impact(df_cleaned)
```

```python
analyze_outlier_impact(df_cleaned)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_location_supply(df_cleaned, matches_path='matches.json'):
    """
    Analyzes the supply of active properties for the non-outlier, zero-match properties.
    """
    try:
        with open(matches_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{matches_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{matches_path}' not found. Please run the matching script first.")
        return

    # --- 1. Identify the non-outlier, zero-match properties ---
    
    # First, find all outliers to exclude them
    numerical_cols = [
        'Property-Price', 'Super-Built-up-Construction-Area', 
        'Carpet-Construction-Area', 'Age-Of-Property', 'BHK', 'Bathroom'
    ]
    outlier_bounds = {}
    for col in numerical_cols:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_bounds[col] = (lower_bound, upper_bound)

    outlier_tags = set()
    for index, row in df_cleaned.iterrows():
        for col in numerical_cols:
            lower, upper = outlier_bounds[col]
            if not (lower <= row[col] <= upper):
                outlier_tags.add(row['Tag'])
                break # Move to the next property once one outlier feature is found

    # Now, find the tags that have zero matches AND are not outliers
    zero_match_tags = {tag for tag, matches in matches_data.items() if len(matches) == 0}
    non_outlier_zero_match_tags = zero_match_tags - outlier_tags
    
    non_outlier_df = df_cleaned[df_cleaned['Tag'].isin(non_outlier_zero_match_tags)]
    
    print(f"Analyzing {len(non_outlier_df)} non-outlier properties with zero matches.")

    if non_outlier_df.empty:
        print("No properties fit this category for analysis.")
        return

    # --- 2. Calculate the supply of active properties in each location cluster ---
    active_properties = df_cleaned[df_cleaned['Calculated_Status'] == 'Active']
    cluster_supply = active_properties['Location_Cluster'].value_counts().to_dict()

    # --- 3. Analyze the supply for our target group ---
    non_outlier_df['Active_Properties_in_Cluster'] = non_outlier_df['Location_Cluster'].map(cluster_supply).fillna(0)
    
    supply_distribution = non_outlier_df['Active_Properties_in_Cluster'].value_counts().sort_index()

    print("\n--- üìä Supply Analysis Results ---")
    print("Distribution of active properties in the clusters of our hard-to-match properties:")
    print(supply_distribution)

    # --- 4. Visualize the findings ---
    plt.figure(figsize=(12, 6))
    sns.barplot(x=supply_distribution.index, y=supply_distribution.values, palette='coolwarm')
    plt.title('Supply of Active Properties for Hard-to-Match Listings')
    plt.xlabel('Number of Active Properties in the Same Cluster')
    plt.ylabel('Number of Hard-to-Match Expired Properties')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# --- How to run this in your notebook ---
# analyze_location_supply(df_cleaned)
```

```python
analyze_location_supply(df_cleaned)
```

```python
import json
import pandas as pd
import numpy as np

def analyze_cluster_specific_outliers(df_cleaned, matches_path='matches.json'):
    """
    Analyzes if the non-outlier, zero-match properties are outliers
    within their own specific location clusters.
    """
    try:
        with open(matches_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{matches_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{matches_path}' not found. Please run the matching script first.")
        return

    # --- 1. Identify the non-outlier, zero-match properties ---
    
    # First, find all global outliers to exclude them
    numerical_cols = [
        'Property-Price', 'Super-Built-up-Construction-Area', 
        'Carpet-Construction-Area', 'Age-Of-Property', 'BHK', 'Bathroom'
    ]
    outlier_bounds = {}
    for col in numerical_cols:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_bounds[col] = (lower_bound, upper_bound)

    outlier_tags = set()
    for index, row in df_cleaned.iterrows():
        for col in numerical_cols:
            lower, upper = outlier_bounds[col]
            if not (lower <= row[col] <= upper):
                outlier_tags.add(row['Tag'])
                break

    zero_match_tags = {tag for tag, matches in matches_data.items() if len(matches) == 0}
    non_outlier_zero_match_tags = zero_match_tags - outlier_tags
    
    non_outlier_df = df_cleaned[df_cleaned['Tag'].isin(non_outlier_zero_match_tags)]
    
    print(f"Analyzing {len(non_outlier_df)} non-outlier properties with zero matches.")

    if non_outlier_df.empty:
        print("No properties fit this category for analysis.")
        return

    # --- 2. Check for cluster-specific outliers ---
    cluster_outlier_properties = []
    cluster_outlier_feature_counts = {col: 0 for col in numerical_cols}

    for index, property_to_check in non_outlier_df.iterrows():
        cluster_id = property_to_check['Location_Cluster']
        
        # Get all other properties in the same cluster
        cluster_df = df_cleaned[df_cleaned['Location_Cluster'] == cluster_id]
        
        is_cluster_outlier = False
        for col in numerical_cols:
            # Calculate the local bounds for this specific cluster
            Q1 = cluster_df[col].quantile(0.25)
            Q3 = cluster_df[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Avoid issues with clusters that have no variance
            if IQR == 0:
                continue
                
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Check if our property is an outlier within its cluster
            if not (lower_bound <= property_to_check[col] <= upper_bound):
                is_cluster_outlier = True
                cluster_outlier_feature_counts[col] += 1
        
        if is_cluster_outlier:
            cluster_outlier_properties.append(property_to_check['Tag'])

    # --- 3. Report the findings ---
    num_cluster_outliers = len(cluster_outlier_properties)
    total_to_check = len(non_outlier_df)
    percentage = (num_cluster_outliers / total_to_check) * 100 if total_to_check > 0 else 0

    print("\n--- üìä Cluster-Specific Outlier Analysis Results ---")
    print(f"Out of {total_to_check} hard-to-match properties, {num_cluster_outliers} ({percentage:.2f}%) were identified as outliers within their own location cluster.")
    
    print("\n--- üî• Most Frequent Cluster-Outlier Features ---")
    print("This shows which features were most often the reason a property was a local outlier:")
    
    sorted_cluster_outlier_features = sorted(cluster_outlier_feature_counts.items(), key=lambda item: item[1], reverse=True)
    for feature, count in sorted_cluster_outlier_features:
        if count > 0:
            print(f"- {feature}: {count} properties")

# --- How to run this in your notebook ---
# analyze_cluster_specific_outliers(df_cleaned)
```

```python
analyze_cluster_specific_outliers(df_cleaned)
```

```python
import json
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from IPython.display import HTML

def generate_unmatched_report(df_cleaned, all_embeddings, matches_path='matches.json'):
    """
    Generates a comprehensive HTML report for ALL hard-to-match properties.
    """
    try:
        with open(matches_path, 'r') as f:
            matches_data = json.load(f)
        print(f"‚úÖ Successfully loaded '{matches_path}'.")
    except FileNotFoundError:
        print(f"‚ùå Error: '{matches_path}' not found. Please run the matching script first.")
        return

    # --- 1. Isolate the hard-to-match properties ---
    numerical_cols = [
        'Property-Price', 'Super-Built-up-Construction-Area', 
        'Carpet-Construction-Area', 'Age-Of-Property', 'BHK', 'Bathroom'
    ]
    outlier_bounds = {}
    for col in numerical_cols:
        Q1 = df_cleaned[col].quantile(0.25); Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR; upper_bound = Q3 + 1.5 * IQR
        outlier_bounds[col] = (lower_bound, upper_bound)

    outlier_tags = set()
    for index, row in df_cleaned.iterrows():
        for col in numerical_cols:
            lower, upper = outlier_bounds[col]
            if not (lower <= row[col] <= upper):
                outlier_tags.add(row['Tag']); break

    zero_match_tags = {tag for tag, matches in matches_data.items() if len(matches) == 0}
    non_outlier_zero_match_tags = list(zero_match_tags - outlier_tags)
    
    if not non_outlier_zero_match_tags:
        print("No non-outlier, zero-match properties to analyze.")
        return
        
    print(f"\n--- üïµÔ∏è‚Äç‚ôÇÔ∏è Generating Root Cause Analysis Report for {len(non_outlier_zero_match_tags)} Properties ---")

    # --- 2. Prepare the kNN model once ---
    active_mask = df_cleaned['Calculated_Status'] == 'Active'
    active_indices = df_cleaned[active_mask].index
    active_embeddings = all_embeddings[active_indices]
    
    knn = NearestNeighbors(n_neighbors=10, metric='cosine')
    knn.fit(active_embeddings)

    # --- 3. Loop through and build the HTML report ---
    html_report_parts = []
    for case_study_tag in non_outlier_zero_match_tags:
        case_study_property = df_cleaned[df_cleaned['Tag'] == case_study_tag].iloc[0]
        
        # --- Part A: Comparative Profile Analysis ---
        query_index = case_study_property.name
        query_embedding = all_embeddings[query_index].reshape(1, -1)
        distances, indices = knn.kneighbors(query_embedding)
        
        comparison_data = [case_study_property]
        for i, index in enumerate(indices[0]):
            match_index = active_indices[index]
            comparison_data.append(df_cleaned.iloc[match_index])
            
        comparison_df = pd.DataFrame(comparison_data)
        
        query_price = case_study_property['Property-Price']
        price_tolerance = 0.30
        lower_bound = query_price * (1 - price_tolerance)
        upper_bound = query_price * (1 + price_tolerance)
        
        def highlight_price(val):
            return f'color: {"red" if not (lower_bound <= val <= upper_bound) else "green"}'

        styled_df = comparison_df[['Tag', 'Property-Price', 'BHK', 'Location_Cluster', 'Carpet-Construction-Area', 'Age-Of-Property']].style.applymap(highlight_price, subset=['Property-Price'])
        
        # --- Part B: Local Cluster Analysis ---
        cluster_id = case_study_property['Location_Cluster']
        cluster_df = df_cleaned[df_cleaned['Location_Cluster'] == cluster_id]
        cluster_stats = cluster_df[numerical_cols].describe().loc[['mean', '50%', 'min', 'max']]
        property_values = case_study_property[numerical_cols]
        property_values.name = 'Expired_Property_Value'
        local_comparison_df = pd.concat([property_values, cluster_stats])
        
        def highlight_local_outlier(row):
            styles = [''] * len(row)
            for i, col in enumerate(numerical_cols):
                val = row[col]
                mean = local_comparison_df.loc['mean', col]
                if val > mean * 1.5 or val < mean * 0.5:
                    styles[i] = 'background-color: #ffcccc'
            return styles

        styled_local_df = local_comparison_df.T.style.apply(highlight_local_outlier, axis=1)

        # --- Append HTML parts for this property ---
        html_report_parts.append(f"<hr style='border: 2px solid #000;'><h2 style='color: #0056b3;'>Analysis for Tag: {case_study_tag}</h2>")
        html_report_parts.append("<h3>Part A: Why Structurally Similar Properties Failed the Price Filter</h3>")
        html_report_parts.append(f"<p>The first row is our EXPIRED property. Price filter requires matches to be between {lower_bound:.2f} and {upper_bound:.2f} Lacs.</p>")
        html_report_parts.append(styled_df.to_html())
        html_report_parts.append("<h3>Part B: Is the Property a 'Local Outlier' in its Neighborhood?</h3>")
        html_report_parts.append(f"<p>Analyzing the property against the {len(cluster_df)} other properties in Location Cluster {cluster_id}.</p>")
        html_report_parts.append(styled_local_df.to_html())

    # --- 4. Save the final HTML report ---
    final_html = f"""
    <html>
    <head>
        <title>Unmatched Properties Analysis Report</title>
        <style>
            body {{ font-family: sans-serif; line-height: 1.6; }}
            table {{ border-collapse: collapse; width: 100%; margin-bottom: 20px; }}
            th, td {{ border: 1px solid #dddddd; text-align: left; padding: 8px; }}
            th {{ background-color: #f2f2f2; }}
        </style>
    </head>
    <body>
        <h1>Root Cause Analysis for Hard-to-Match Properties</h1>
        {''.join(html_report_parts)}
    </body>
    </html>
    """
    
    report_path = 'unmatched_properties_report.html'
    with open(report_path, 'w') as f:
        f.write(final_html)
        
    print(f"\n‚úÖ Complete analysis report has been saved to '{report_path}'.")
    print("You can now open this file in your browser and print it to PDF.")

# --- How to run this in your notebook ---
# generate_unmatched_report(df_cleaned, all_embeddings)
```
